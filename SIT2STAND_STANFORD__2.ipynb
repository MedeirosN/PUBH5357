```r
# Load necessary libraries
library(readr)

# Load the dataset (update the path if necessary)
data <- read_csv("dataClean.csv")

# Remove non-numeric columns
data_numeric <- data[sapply(data, is.numeric)]

# Remove rows with missing values
data_numeric <- na.omit(data_numeric)

# Ensure there are enough variables for regression
if (ncol(data_numeric) > 1) {

  # Define the regression formula using all variables
  formula <- as.formula(paste(names(data_numeric)[1], "~", paste(names(data_numeric)[-1], collapse = " + ")))

  # Fit the regression model
  model <- lm(formula, data = data_numeric)

  # Save the summary output to a file
  summary_output <- capture.output(summary(model))
  writeLines(summary_output, "regression_summary.txt")

  # Print a success message
  print("Regression analysis complete. Results saved to 'regression_summary.txt'")

} else {
  print("Not enough numeric columns to perform regression.")
}
```

```r
# Function to classify correlation strength
classify_correlation <- function(value) {
  if (value == 1) return("Perfect Positive")
  if (value >= 0.7) return("Strong Positive")
  if (value >= 0.5) return("Moderate Positive")
  if (value >= 0.3) return("Weak Positive")
  if (value > -0.3) return("No Correlation")
  if (value >= -0.5) return("Weak Negative")
  if (value >= -0.7) return("Moderate Negative")
  if (value >= -1) return("Strong Negative")
  return("Perfect Negative")
}

# Apply classification
correlation_results <- data.frame(
  Variable = dependent_vars,
  Correlation_with_time_sit2stand = correlations_time_sit2stand,
  Classification_time_sit2stand = sapply(correlations_time_sit2stand, classify_correlation),
  Correlation_with_speed_stand2sit = correlations_speed_stand2sit,
  Classification_speed_stand2sit = sapply(correlations_speed_stand2sit, classify_correlation)
)

# Display the correlation table
print(correlation_results)

# Save results to CSV file
write.csv(correlation_results, "correlation_results.csv", row.names = FALSE)

# Generate correlation heatmap
cor_matrix <- cor(data[, c(independent_vars, dependent_vars)], use = "complete.obs")

# Plot correlation heatmap using ggcorrplot
ggcorrplot(cor_matrix, method = "square", type = "lower", lab = TRUE, lab_size = 3,
           colors = c("blue", "white", "red"), title = "Correlation Heatmap", ggtheme = theme_minimal())

# Bar charts for correlation values
ggplot(correlation_results, aes(x = reorder(Variable, Correlation_with_time_sit2stand), y = Correlation_with_time_sit2stand, fill = Classification_time_sit2stand)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  ggtitle("Correlation with Time Sit-to-Stand") +
  xlab("Variables") +
  ylab("Correlation") +
  theme_minimal()

ggplot(correlation_results, aes(x = reorder(Variable, Correlation_with_speed_stand2sit), y = Correlation_with_speed_stand2sit, fill = Classification_speed_stand2sit)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  ggtitle("Correlation with Speed Stand-to-Sit") +
  xlab("Variables") +
  ylab("Correlation") +
  theme_minimal()
```
